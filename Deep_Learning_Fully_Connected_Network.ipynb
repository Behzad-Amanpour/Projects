#Dataset

```python
from sklearn.datasets import load_breast_cancer
# X, y = load_breast_cancer( return_X_y = True )
X = load_breast_cancer().data
y = load_breast_cancer().target
y = (-1*y) + 1

from sklearn.preprocessing import MinMaxScaler
X = MinMaxScaler().fit( X ).transform( X )

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=21 )

from tensorflow import one_hot
Num_Classes = 2
train_labels = one_hot(y_train, Num_Classes)
test_labels = one_hot(y_test, Num_Classes)
```

```python
import keras
from keras import layers
from keras.applications import EfficientNetB0

inputs = layers.Input(shape=(224, 224, 3))

base_model = EfficientNetB0(include_top=False, input_tensor=inputs, weights="imagenet") #include_top=False:

# Freeze the pretrained weights
base_model.trainable = False

# Rebuild top
x = layers.GlobalAveragePooling2D(name="avg_pool")(base_model.output) #Average of feature map.
x = layers.BatchNormalization()(x)

top_dropout_rate = 0.2

x = layers.Dropout(top_dropout_rate, name="top_dropout")(x)
#or x=layers.Dropuot(rate = 0.2)(x)

outputs = layers.Dense(2, activation="softmax", name="pred")(x)

model = keras.Model(inputs, outputs, name="EfficientNet")

############
!pip install visualkeras
import visualkeras
from collections import defaultdict
color_map = defaultdict(dict)

color_map[layers.GlobalAveragePooling2D]['fill'] = 'red'
color_map[layers.BatchNormalization]['fill'] = 'black'
color_map[layers.Dropout]['fill'] = 'yellow'
color_map[layers.Dense]['fill'] = 'green'

#visualkeras.layered_view(model, color_map=color_map, legend=True,
#                         show_dimension=True, to_file='output.png',
#                         draw_reversed=True, spacing=20) #draw_reversed=True/Flase (we can look at the back of each layer.)

visualkeras.layered_view(model, color_map=color_map, legend=True, show_dimension=True, draw_reversed= True,
                         scale_xy=10, scale_z=10, max_z=80, spacing=30)
```

```python
#https://gist.github.com/craffel/2d727968c3aaebd10359

import matplotlib.pyplot as plt

def draw_neural_net(ax, left, right, bottom, top, layer_sizes):

    n_layers = len(layer_sizes)
    v_spacing = (top - bottom)/float(max(layer_sizes))
    h_spacing = (right - left)/float(len(layer_sizes) - 1)
    # Nodes
    for n, layer_size in enumerate(layer_sizes):
        layer_top = v_spacing*(layer_size - 1)/2. + (top + bottom)/2.
        for m in range(layer_size):
            circle = plt.Circle((n*h_spacing + left, layer_top - m*v_spacing), v_spacing/4.,
                                color='w', ec='k', zorder=4)
            ax.add_artist(circle)
    # Edges
    for n, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
        layer_top_a = v_spacing*(layer_size_a - 1)/2. + (top + bottom)/2.
        layer_top_b = v_spacing*(layer_size_b - 1)/2. + (top + bottom)/2.
        for m in range(layer_size_a):
            for o in range(layer_size_b):
                line = plt.Line2D([n*h_spacing + left, (n + 1)*h_spacing + left],
                                  [layer_top_a - m*v_spacing, layer_top_b - o*v_spacing], c='k')
                ax.add_artist(line)


fig = plt.figure(figsize=(3, 3))
ax = fig.gca()
ax.axis('off')
draw_neural_net(ax, .1, .9, .1, .9, [4, 7,2])
#fig.savefig('nn.png')
```

#Simple Training

##Training

```python
# optimizer = keras.optimizers.Adam(learning_rate=1e-2)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

epochs = 20

batch_size = 4

hist = model.fit(
                 x = X_train,
                 y = train_labels,
                 batch_size = batch_size,
                 epochs = epochs,
                 # callbacks= [CustomCallback(x)], # None, augmentation
                 validation_split=0.2,
                 # validation_data=(valid_images, valid_labels),
                 shuffle=True,
                 )
```

##Test

```python
from sklearn.metrics import classification_report

predictions = model.predict(X_test)

# Convert one-hot encoded labels to class labels (0 or 1)
predicted_labels = np.argmax(predictions, axis=1)

print("\nClassification Report:\n", classification_report(y_test, predicted_labels))
```
